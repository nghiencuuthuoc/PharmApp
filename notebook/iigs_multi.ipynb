```python
import sys
sys.path.append('..')
import os
# os.getcwd()
import pandas as pd
import numpy as np
import pickle
import shutil
from pandasgui import show
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)




# # load pickle but error
# iigs_pickle_path = '../pickle/inactive_ingredients_sagerx.pickle'
# if os.path.isfile(iigs_pickle_path) == True:
#     print('pickle file exist!')
#     inactive_ingredients_sagerx = pd.read_pickle(iigs_pickle_path)
#     df_pickle = inactive_ingredients_sagerx
# else:
#     inactive_ingredients_sagerx = pd.read_csv("../sagerxdata/products_to_inactive_ingredients.csv", low_memory=False)
#     inactive_ingredients_sagerx.to_pickle(iigs_pickle_path)
#     print('pickle file saved!')
#     inactive_ingredients_sagerx = pd.read_pickle(iigs_pickle_path)
#     df_pickle = inactive_ingredients_sagerx

# input name
drug_input = str(input("Enter name: "))
# drug_name = 'nexium'
drug_name = drug_input.lower().replace(' ', '+').replace('-','+')

# Save file iigdata
iig_file = drug_name + "_inactive_ingredient_list.txt"
iig_file_path = os.path.join('../iigdata', iig_file)
if os.path.isfile(iig_file_path) == True:
    print(f'{iig_file} exist in iigdata folder!')
else:
    # get iig
    # import src.pharmappiig013 as pa
    import src.pharmappiig017dev as pa
    pa.get_inactive_ingredient(name_drug=drug_name)


def search(df: pd.DataFrame, substring: str, case: bool = False) -> pd.DataFrame: 
    mask = np.column_stack([df[col].astype(str).str.contains(substring.lower(), case=case, na=False) for col in df])
    return df.loc[mask.any(axis=1)]

# Save file iigsdata
iigs_file = drug_name + "_iigs.txt"
iigs_file_path = os.path.join('../iigsdata', iigs_file)
if os.path.isfile(iigs_file_path) == True:
    print(f'{iigs_file} exist in iigsdata folder!')
else:
    inactive_ingredients_sagerx = pd.read_csv("../sagerxdata/products_to_inactive_ingredients.csv", low_memory=False)
    df_pickle = inactive_ingredients_sagerx  
    df_iigs = search(df_pickle, drug_name)
    df_iigs.to_csv(iigs_file_path, mode='a', index = False, header=True)
    print(f'{iigs_file} saved in iigsdata folder!')


# delete all file in iigview
iigview_folder = '../iigview'
iigview_list_file = os.listdir(iigview_folder)
for item in iigview_list_file:
    if item.endswith(".txt"):
        os.remove(os.path.join(iigview_folder, item))

# Save iigs to excel
iigs_file = drug_name + "_iigs.txt"
iigs_file_path = os.path.join('../iigsdata', iigs_file)

if os.path.isfile(iigs_file_path) == True:
    shutil.copy(iigs_file_path, '../iigview')
    import pandas as pd
    pd.set_option("display.max_rows", None, "display.max_columns", None)
    df_iigs = pd.read_csv(iigs_file_path)
    
    # export to excel
    if df_iigs is not None:
        iig_file_xlsx = drug_name + '_iig.xlsx'
        name_sheet = drug_name + '_iigs'
        iig2excel_folder = '../iig2excel'
        path_file_excel_iigs = iig2excel_folder +'/' + iig_file_xlsx
        with pd.ExcelWriter(path_file_excel_iigs) as writer:
            df_iigs.to_excel(writer, sheet_name = name_sheet)
        # os.system('start "excel" path_file_excel_iigs')
        # path_file_excel_iigs_open = 'D:\\PharmApp\\iig2excel'
        # full_path_file_excel_iigs_open = path_file_excel_iigs_open + '\\' + iig_file_xlsx
        # os.startfile(full_path_file_excel_iigs_open)
else:
    print(f'The {df_iigs} not exist in ../iigsdata!')



# Save iig to excel 
import os
iig_file = drug_name + "_inactive_ingredient_list.txt"
iig_file_path = os.path.join('../iigdata', iig_file)


if os.path.isfile(iig_file_path):
    shutil.copy(iig_file_path, '../iigview')
    import pandas as pd
    pd.set_option("display.max_rows", None, "display.max_columns", None)
    df_iig = pd.read_csv(iig_file_path)
    
    # export to excel
    if df_iig is not None:
        iig_file_xlsx = drug_name + '_iig.xlsx'
        name_sheet = drug_name + '_iig'
        iig2excel_folder = '../iig2excel'
        path_file_excel_iig = iig2excel_folder +'/' + iig_file_xlsx
        with pd.ExcelWriter(path_file_excel_iig, engine='openpyxl', mode='a') as writer:
            df_iig.to_excel(writer, sheet_name = name_sheet)
        # os.system('start "excel" path_file_excel_iig')
        path_file_excel_iig_open = 'D:\\PharmApp\\iig2excel'
        full_path_file_excel_iig_open = path_file_excel_iig_open + '\\' + iig_file_xlsx
        os.startfile(full_path_file_excel_iig_open)
else:
    print(f'Inactive ingredient {drug_name} not exist in ../iigdata!')

# view multi drug with pandasgui
folder_path = '../iigview'
txt_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

dataframes = {}
for txt_file in txt_files:
    file_path = os.path.join(folder_path, txt_file)
    df = pd.read_csv(file_path, delimiter=',')  
    key = os.path.splitext(txt_file)[0]
    key = key.replace('_inactive_ingredient_list','_iig').replace('+', ' ')
    dataframes[key] = df
  
for key, df in dataframes.items():
    print(f"DataFrame for {key}:")
    # print(df.head())
gui = show(**dataframes)
```

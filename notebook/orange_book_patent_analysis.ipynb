````python
# search df
def search_df(df: pd.DataFrame, substring: str, case: bool = False) -> pd.DataFrame: 
    mask = np.column_stack([df[col].astype(str).str.contains(substring.lower(), case=case, na=False) for col in df])
    return df.loc[mask.any(axis=1)]
````

````python
df_dict = {}
merging_indices=['appl_no', 'product_no']
for root, dirs, files in os.walk('../orange_book/raw_data/'):
    for file in files:
        filename, extension = os.path.splitext(file)
        if extension == '.txt':
            df_dict[filename] = pd.read_csv(os.path.join(root, file), sep='~', engine='python')
            df_dict[filename].columns = map(str.lower, df_dict[filename].columns) #lowercase column names
# for k, v in df_dict.items():
#     print(k, '\n', '-----------')
#     print(v.info())
#     print('\n'*2)
#Merging patent datasets
orange_book_full = pd.merge(df_dict['products'], df_dict['patent'], on=merging_indices, how='left')
orange_book_full = pd.merge(orange_book_full, df_dict['exclusivity'], on=merging_indices, how='left')
````
